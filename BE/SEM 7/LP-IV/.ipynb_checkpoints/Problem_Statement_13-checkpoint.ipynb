```python
# Cell 1 - Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import SGD
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Cell 2 - Load CSVs (adjust paths if needed)
# Assumes CSV has header row with pixel_0...pixel_3071 and last column named 'label' (or similar)
train = pd.read_csv('datasets/cifr/train_data.csv', header=0)
test  = pd.read_csv('datasets/cifr/test_data.csv',  header=0)
```

```python
# Cell 3 - Prepare X and y (drop last column as label)
X_train = train.iloc[:, :-1].to_numpy().astype('float32')
y_train = train.iloc[:, -1].to_numpy().astype('int32')
X_test  = test.iloc[:, :-1].to_numpy().astype('float32')
y_test  = test.iloc[:, -1].to_numpy().astype('int32')

print("raw shapes:", X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# Cell 4 - Reshape into images and normalize
# CIFAR images: 32 x 32 x 3 -> total 3072 pixels
X_train = X_train.reshape((-1, 32, 32, 3)) / 255.0
X_test  = X_test.reshape((-1, 32, 32, 3))  / 255.0

print("image shapes:", X_train.shape, X_test.shape)
```

```python
# Cell 5 - Build a simple CNN (Sequential API)
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    MaxPooling2D((2,2)),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(64, (3,3), activation='relu'),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer=SGD(learning_rate=0.01), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# Cell 6 - Train the model (SGD, 11 epochs as requested)
epochs = 11
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=128, verbose=1)
```

```python
# Cell 7 - Evaluate + Reports + Plots
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}   Test Accuracy: {acc*100:.2f}%")

# Predictions and classification report / confusion matrix
y_pred = model.predict(X_test).argmax(axis=1)
print("\nClassification report:\n", classification_report(y_test, y_pred, digits=4))
print("Confusion matrix shape:", confusion_matrix(y_test, y_pred).shape)
```

```python
# Plot training curves (loss & accuracy)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Accuracy'); plt.xlabel('epoch'); plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss'); plt.xlabel('epoch'); plt.legend()
plt.show()
```
